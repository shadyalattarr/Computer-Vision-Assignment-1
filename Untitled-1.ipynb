```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # can be used to make histogram of the image
import cv2
```

```python
print(cv2.__version__)
```

```python
# cv2.imread(r"path") method loads an image from the specified file -- read in BGR intead of RGB
# cv2.IMREAD_COLOR loads a color image
# cv2.IMREAD_GRAYSCALE loads image in grayscale mode
# cv2.IMREAD_UNCHANGED loads image as such including alpha channel
img = cv2.imread(r'images\\img1.png')
```

```python
pd.Series(img.flatten()).plot(kind='hist', bins=50, figsize=(10, 5), title='Distribution of Pixel Values')
# flatten() method returns a copy of the array collapsed into one dimension
# pd.Series() method converts the array into a pandas series
plt.show()
```

```python
# Display the image
cv2.imshow('start image', img)
cv2.waitKey(0) # to auto close ig
cv2.destroyAllWindows()
# waitKey() methkod waits for a specified amount of time (in milliseconds) for a key stroke
# destroyAllWindows() method closes all windows and de-allocates any associated memory usage
```

```python
imgrgb = cv2.imread(r'images\\rgb.png')
cv2.imshow('rgb image', imgrgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to show specific channels
cv2.imshow('red channel', imgrgb[:, :, 2]) # BGR
cv2.waitKey(0)
cv2.destroyAllWindows()
# it is black and white because it is only one channel
# black in areas where there is no red
# there is red in the red circle and in white parts
```

```python
cv2.imshow('green channel', img[:, :, 1]) # BGR

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# can read using plt too
img_plt = plt.imread(r'images\\img1.png')
fig, ax = plt.subplots(1, 3, figsize=(15, 5)) # 1 row three cols
ax[0].imshow(img_plt)
ax[0].set_title('Using plt')
ax[1].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # convert BGR to RGB
ax[1].set_title('Using cv2 with conversion')
ax[2].imshow(img)
ax[2].set_title('Using cv2 no conversion')
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
plt.show()  
# looks weird because of BGR
```

```python
rgb_plt = plt.imread(r'images\\rgb.png')
fig, ax = plt.subplots(1, 4, figsize=(15, 5))
ax[0].imshow(rgb_plt[:, :, 0], cmap='Reds') # cmap for visualization
ax[0].set_title('red channel')
ax[1].imshow(rgb_plt[:, :, 1], cmap='Greens')
ax[1].set_title('green channel')
ax[2].imshow(rgb_plt[:, :, 2], cmap='Blues')
ax[2].set_title('blue channel')
ax[3].imshow(rgb_plt)
ax[3].set_title('original image')
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
ax[3].axis('off')
plt.show()
```
