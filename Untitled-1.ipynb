### convert to gray

```python
# Convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the image using Matplotlib
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# cv2.imshow("Grayscale Image", gray_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

## Edits

```python
# median
median_img = cv2.medianBlur(gray_img, 5)  # I think its different from pdf 

# plt.figure(figsize=(10, 5))
# original gray image
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
#image after median blur
plt.subplot(1, 2, 2)
plt.imshow(median_img, cmap='gray')
plt.title("Filtered Image (Median Blur)")
plt.axis('off')

plt.show()
```

```python
# laplacian
laplacian_img = cv2.Laplacian(median_img , cv2.CV_64F,ksize = 5)  # cv2.CV_64F is a signed float64

laplacian_img = np.uint8(np.absolute(laplacian_img)) # convert to unsigned int to display it correctly
# since cv2.imshow() expects uint8 images

# plt.figure(figsize=(10, 5))
# original gray image
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
#image after laplacian filter
plt.subplot(1, 2, 2)
plt.imshow(laplacian_img, cmap='gray')
plt.title("Filtered Image (Laplacian)")
plt.axis('off')
```

## part 2

```python
img_cv2 = cv2.imread(r'images\\Road_img.png')
img_plt = plt.imread(r'images\\Road_img.png')
plt.imshow(img_plt)
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
gray_img = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2GRAY)

# Display the image using Matplotlib
plt.imshow(gray_img, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

```python
# median
median_img = cv2.medianBlur(gray_img, 5)  

# Display the images
# original gray image
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
#image after median blur
plt.subplot(1, 2, 2)
plt.imshow(median_img, cmap='gray')
plt.title("Filtered Image (Median Blur)")
plt.axis('off')

plt.show()
```

```python
edges = cv2.Canny(median_img, threshold1=100, threshold2=200)

# Display the results
# cv2.imshow("Original Image", img_cv2)
# cv2.imshow("Canny Edges", edges)

plt.subplot(1, 2, 1)
plt.imshow(img_plt)  # Show original image in RGB
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")  # Show Canny edges in grayscale
plt.title("Canny Edges")
plt.axis("off")

plt.tight_layout()
plt.show()
```

#### from here

```python
# Define the coordinates of the ROI (Face region)
x1, y1 = 100, 50    # Top-left corner
x2, y2 = 300, 200   # Bottom-right corner

# Extract ROI
roi = img_plt[y1:y2, x1:x2]  # Select the region of interest

# Display Original and Extracted ROI
# plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_plt)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(roi)
plt.title("Extracted ROI (Face)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=50, maxLineGap=100)
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img_cv2, (x1, y1), (x2, y2), (0, 255, 0), 3)
cv2.imshow("Hough Lines", img_cv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
